```python
df = pd.read_csv('insurance2.csv')
```

```python
#Criando uma aleatória de resposta
#0= não comprou; 1= comprout
```

```python
df['ID'] = df.reset_index().astype('int32')
df
```

```python
#alteração do sexo
df['sex_classification'] = df['sex'].apply(lambda x: 'Feminino' if x==0 else 'Masculino')
df
```

```python
df['Fumante'] = df['smoker'].apply(lambda x: 'Não Fumante' if x==0 else 'Fumante')
df
```

```python
#Usando apenas o que reivindicaram
variavel_filtrada =  df['insuranceclaim'] == 1
variavel_filtrada = df[variavel_filtrada]
variavel_filtrada
```

```python
#quantidade de reinvidicação do seguro por genero
revindicaçao_genero = variavel_filtrada.groupby('sex_classification').count()[['insuranceclaim']].reset_index()
revindicaçao_genero
```

```python
#quantidade de reinvidicação do seguro por fumantes
revindicaçao_fumantes = df.groupby('Fumante').count()[['insuranceclaim']].reset_index()
revindicaçao_fumantes
```

```python
#media de custo por genero
revindicaçao_custo = df.groupby('sex_classification').mean()[['charges']].reset_index()
revindicaçao_custo
```
